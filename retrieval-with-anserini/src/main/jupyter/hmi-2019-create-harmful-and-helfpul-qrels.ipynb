# Create harmful-only and helpful-only qrels for 2019

```python
import pandas as pd

DIR = '../../../../third-party/health-misinfo-19/'
CORRECT = 1
INCORRECT = 0

df_corr = pd.read_csv(DIR + '/2019qrels_correctness.txt', sep=' ', names=['Topic', 'Q0', 'DocId', 'Relevance', 'Correctness', 'Credibility'])
df_corr
```

```python
df_helpful = df_corr[(df_corr['Relevance'] > 0) & (df_corr['Correctness'] == CORRECT)].copy()

del df_helpful['Correctness']
del df_helpful['Credibility']

df_helpful
```

```python
df_harmful = df_corr[(df_corr['Relevance'] > 0) & (df_corr['Correctness'] == INCORRECT)].copy()

del df_harmful['Correctness']
del df_harmful['Credibility']

df_harmful
```

```python
df_helpful.to_csv(DIR + 'misinfo-qrels-graded.helpful-only', sep=' ', index=False, header=False)
df_harmful.to_csv(DIR + 'misinfo-qrels-graded.harmful-only', sep=' ', index=False, header=False)
```
